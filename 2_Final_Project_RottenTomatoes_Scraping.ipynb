```python
from bs4 import BeautifulSoup
import requests
print('requests imported')
```

```python
#name the url as the webpage we would like to scrape from IMDB
#this url is about all the Marvel Studio Movies
url = 'https://www.rottentomatoes.com/franchise/marvel_cinematic_universe'
resp = requests.get(url)
```

```python
#turn the request into text file
soup = BeautifulSoup(resp.text)
soup
```

```python
#find all the movie information
#we can see tht all the movie info are stored under the div 'liter-item-content'
movie_titles = soup.find_all('div', class_ = 'media franchiseItem')
```

```python
print(type(movie_titles))
print(len(movie_titles))
```

```python
# Let's figure stuff out first for the second movie (1 position) 
# we didn't use the 0 position since Avengers:Endgame doesn't have ratings yet
first_movie = movie_titles[1]
first_movie
```

```python
# Movie names are under "<a href=" under the "strong" section, so we'll test the following
first_movie.strong.a
```

```python
# Extract the name of the movie as follows
first_name = first_movie.strong.a.text
first_name
```

```python
# Next, we want to find the Rotten Tomatoe rating of Captain Marvel the rating info is store under div
first_rating = first_movie.find('span', class_ = 'meter-value').text.replace('\n', '').replace(' ', '') 
#first_rating = first_rating.text.replace('\n', '').replace(' ', '')
first_rating
```

```python
first_year = first_movie.find('span', class_='subtle').text.replace('(','').replace(')','')
first_year
```

```python
movie_name = []
Year_Released = []
RT = []
```

```python
for i in movie_titles:
    if i.find('span', class_ = 'meter-value') is not None:
        name = i.strong.a.text
        movie_name.append(name)
        
        years = i.find('span', class_='subtle').text.replace('(','').replace(')','')
        Year_Released.append(years)
        
        rt = i.find('span', class_ = 'meter-value').text.replace('\n', '').replace(' ', '') 
        RT.append(rt)
        
        
        
        
```

```python
import pandas as pd
test_df = pd.DataFrame({'Movie': movie_name,
                       'Year_Released': Year_Released,
                       'RT':RT})
```
