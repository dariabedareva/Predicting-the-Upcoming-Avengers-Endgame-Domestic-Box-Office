```python
from bs4 import BeautifulSoup
import requests
print('requests imported')
```

```python
#name the url as the webpage we would like to scrape from IMDB
#this url is about all the Marvel Studio Movies
url = 'https://www.imdb.com/list/ls026690821/'
resp = requests.get(url)
```

```python
#turn the request into text file
soup = BeautifulSoup(resp.text)
soup
```

We want to find all the movies' names
so we explore the url by right-click inspect on the page and found that the movie titles are stored under 'h3'


```python
#find all the movie information
#we can see tht all the movie info are stored under the div 'liter-item-content'
movie_titles = soup.find_all('div', class_ = 'lister-item-content')
```

```python
print(type(movie_titles))
print(len(movie_titles))
```

```python
first_movie = movie_titles[0]
first_movie
```

The data in 'div' is still very long, so we go back to the website again and found that the movie names are store under'h3'

```python
#extract 'Iron Man'
first_name = first_movie.h3.a.text
first_name
```

```python
#now we want to find the year that the movie release. we return to the url page and found the year is stored as"<span class="lister-item-year text-muted unbold">(2008)</span>"
first_year = first_movie.h3.find('span', class_ = 'lister-item-year text-muted unbold')
first_year = first_year.text[1:5]
first_year
```

```python
#Next, we want to find the IMDB rating of Iron Man, the rating info is store under div
first_rate = first_movie.div.find('span', class_ = 'ipl-rating-star__rating').text
first_rate = float(first_rate)
first_rate
```

```python
#scrape the Metascrore of Iron Man, the data is now stored underdiv 
first_votes = first_movie.find('span', attrs = {'name':'nv'})
first_v = first_votes['data-value']
```

```python
#treat the value into a integar
first_votes = int(first_v)
first_votes
```

```python
movie_name = []
years = []
imdb_ratings = []
metascore = []
votes = []
gross = []
```

```python
none_mscore = movie_titles[21].find('div', class_= 'ratings-metascore')
type(none_mscore)
```

```python
for i in movie_titles:
    if i.find('div', class_= 'ratings-metascore') is not None:

        # The name 
        #first_name = first_movie.h3.a.text
        name = i.h3.a.text
        movie_name.append(name)

        # The year 
        #first_year = first_movie.h3.find('span', class_ = 'lister-item-year text-muted unbold')
        #first_year = first_year.text
        year = i.h3.find('span', class_ = 'lister-item-year text-muted unbold').text
        years.append(year)

        # The IMDB rating
        #first_rate = first_movie.div.find('span', class_ = 'ipl-rating-star__rating')
        #first_rate = first_rate.text
        rates = i.div.find('span', class_= 'ipl-rating-star__rating').text
        imdb_ratings.append(float(rates))

        # The Metascore
        #first_ms = first_movie.find('span', class_= 'metascore favorable' )
        #first_ms = int(first_ms.text)
        m_score = i.find('span', class_ = 'metascore').text
        metascore.append(int(m_score))

        # The number of votes
        #first_votes = first_movie.find('span', attrs = {'name':'nv'})
        #first_votes
        #first_votes['data-value']
        vote = i.find('span', attrs = {'name':'nv'})['data-value']
        votes.append(int(vote))
        
        #The number of gross
        #first_gross = first_movie.findAll('span', attrs = {'name':'nv'})[1]
        #first_gross['data-value']
        #first_g = locale.atoi(first_gross)
        grossv = i.find_all('span', attrs = {'name':'nv'})[1]['data-value']
        grossv =locale.atoi(grossv)
        gross.append(int(grossv))
    
```
